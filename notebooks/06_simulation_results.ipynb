## 1. Table 13.1: System Latency vs Semantic Ambiguity

```python
sim = AIoTWorkSystemSimulation()
print('Running simulation...')
results_df = sim.run_ambiguity_sweep(
    ambiguity_levels=[0.0, 0.1, 0.2, 0.3, 0.4],
    iterations=100
)
results_df
```

```python
# Visualize latency vs ambiguity
fig, ax = plt.subplots(figsize=(12, 6))
x = results_df['Semantic Ambiguity (ψ)']
y = results_df['Avg System Latency (sec)'].replace([np.inf], 5000)

ax.plot(x, y, 'o-', linewidth=2, markersize=10, color='#2E86AB')
ax.axhline(y=1000, color='red', linestyle='--', linewidth=2, label='Critical Threshold')
ax.set_xlabel('Semantic Ambiguity (ψ)', fontsize=12)
ax.set_ylabel('Average System Latency (seconds)', fontsize=12)
ax.set_title('System Latency vs Semantic Ambiguity (Equation 12.3)', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../results/figures/sim_latency_vs_ambiguity.png', dpi=300)
plt.show()
```
