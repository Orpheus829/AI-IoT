Test Lesk, Embedding-based, and Ensemble WSD methods.

```python
import sys
sys.path.append('../modules')
import numpy as np
import pandas as pd
from semantic_disambiguation import IntegratedWSDSystem, INDUSTRIAL_SENSE_INVENTORY, INDUSTRIAL_GLOSSES, LeskAlgorithm
```

## 1. Test WSD System

```python
wsd = IntegratedWSDSystem(INDUSTRIAL_SENSE_INVENTORY, INDUSTRIAL_GLOSSES)

test_commands = [
    'check the assembly line',
    'stop the robot arm immediately',
    'reduce load on electrical line',
    'inspect the hydraulic line pressure',
    'check system status'
]

results = []
for cmd in test_commands:
    result = wsd.process_command(cmd)
    results.append({
        'Command': cmd,
        'Confidence': result['avg_confidence'],
        'Ambiguity': result['max_ambiguity'],
        'Disambiguated': str(result['disambiguated'])
    })

pd.DataFrame(results)
```

## 2. Compare WSD Methods

```python
# Test different methods
for method in ['lesk', 'embedding', 'ensemble']:
    result = wsd.disambiguate('line', 'check the assembly line', method=method)
    print(f'{method.upper()}: sense={result["sense"]}, confidence={result["confidence"]:.2f}, ambiguity={result["ambiguity"]:.2f}')
```
