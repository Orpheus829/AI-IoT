Apply Kalman filtering for sensor fusion and noise reduction.

```python
import sys
sys.path.append('../modules')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from kalman_filter import create_machine_health_kf, create_temperature_kf, MultiSensorFusion
```

## 1. Load Processed Data

```python
df = pd.read_csv('../data/ai4i2020_processed.csv')
print(f'Loaded {len(df)} records')
df.head()
```

## 2. Kalman Filter for Tool Wear

```python
kf = create_machine_health_kf()
measurements = df['tool_wear'].values[:200]

filtered = []
for z in measurements:
    x_est = kf.filter(z.reshape(1))
    filtered.append(x_est[0])

plt.figure(figsize=(14, 6))
plt.plot(measurements, alpha=0.5, label='Raw measurements', linewidth=1)
plt.plot(filtered, label='Kalman filtered', linewidth=2)
plt.xlabel('Sample')
plt.ylabel('Tool Wear')
plt.title('Kalman Filter: Tool Wear Estimation')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f'Noise reduction: {(1 - np.std(filtered)/np.std(measurements))*100:.1f}%')
```
