## 1. Hick-Hyman Law

```python
hh = HickHymanModel()

# Test with different ambiguity levels
for n_choices in [2, 4, 8]:
    clean = hh.decision_time(n_choices, ambiguity=0.0)
    ambiguous = hh.decision_time(n_choices, ambiguity=0.3)
    print(f'Choices={n_choices}: Clean={clean:.3f}s, Ambiguous={ambiguous:.3f}s, Increase={((ambiguous/clean-1)*100):.1f}%')
```

## 2. Yerkes-Dodson Law

```python
yd = YerkesDodsonModel()
stress_levels = np.linspace(0, 1, 100)
performance = [yd.performance(s) for s in stress_levels]

plt.figure(figsize=(10, 6))
plt.plot(stress_levels, performance, linewidth=2)
plt.axvline(x=0.5, color='red', linestyle='--', label='Optimal stress')
plt.xlabel('Stress Level')
plt.ylabel('Performance')
plt.title('Yerkes-Dodson Law: Stress vs Performance')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
