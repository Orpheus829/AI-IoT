Markov chains, Weibull reliability, and queueing models.

```python
import sys
sys.path.append('../modules')
import numpy as np
import matplotlib.pyplot as plt
from stochastic_engine import create_machine_health_markov, WeibullReliability, QueueingModel, PoissonProcess
```

## 1. Markov Chain Analysis

```python
mc = create_machine_health_markov()
steady = mc.steady_state()
print('Steady-state probabilities:')
for i, prob in enumerate(steady):
    print(f'  {mc.state_names[i]}: {prob:.4f}')

# Simulate path
path = mc.simulate(n_steps=100, initial_state=0)
print(f'\nSimulated path (first 20): {path[:20]}')
```

## 2. Weibull Reliability

```python
weibull = WeibullReliability(shape=2.5, scale=1000)
times = np.linspace(0, 2000, 100)
reliability = [weibull.reliability(t) for t in times]

plt.figure(figsize=(12, 5))
plt.plot(times, reliability, linewidth=2)
plt.xlabel('Time (hours)')
plt.ylabel('Reliability R(t)')
plt.title('Weibull Reliability Function')
plt.grid(True, alpha=0.3)
plt.show()

print(f'MTTF: {weibull.mean_time_to_failure():.0f} hours')
```

## 3. Queueing Analysis

```python
queue = QueueingModel(arrival_rate=40, service_rate=60)
print(f'Utilization: {queue.utilization():.2f}')
print(f'Mean queue length: {queue.mean_queue_length():.2f}')
print(f'Mean waiting time: {queue.mean_waiting_time():.4f} hours')
```
